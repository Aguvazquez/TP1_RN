```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import visualkeras
import tensorflow as tf
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
```

```python

x_train = np.load("Data/train_images.npy")
y_train = pd.read_csv("Data/train_labels.csv")
x_train = x_train/255
```

```python
#Mean and Variance of all clases.
all_clases_mean = np.zeros([10,784])
all_clases_var = np.zeros([10,784])
for i in range(1,11):
    s_train = x_train[y_train.label.values == i-1, ...] 
    all_clases_mean[i-1] = s_train.mean(axis=0).flatten()
    all_clases_var[i-1] = s_train.std(axis=0).flatten()
    plt.subplot(5,4,2*i - 1)
    a = plt.imshow(s_train.mean(axis=0))
    #plt.colorbar(a)
    plt.subplot(5,4,2*i)
    b = plt.imshow(s_train.std(axis=0))
    #plt.colorbar(b)
```

```python
#Histograma de intensidades.
label_dict = {  0: "T-shirt/top",
                1: "Trouser",
                2: "Pullover",
                3: "Dress",
                4: "Coat",
                5: "Sandal",
                6: "Shirt",
                7: "Sneaker",
                8: "Bag",
                9: "Ankle boot"
             }
for i in range(10):
    plt.hist(all_clases_mean[i] , bins=50,label = label_dict[i],alpha = 0.4)
plt.legend()
plt.title("Histograma de instensidad")
plt.xlabel("Intensidad normalizada")
plt.ylabel("Repeticiones")
plt.grid(True)
```

```python
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
#x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
# Define the ImageDataGenerator and set the augmentation parameters
datagen1 = ImageDataGenerator(
    rotation_range=0,
    width_shift_range=0,
    height_shift_range=0.18,
    zoom_range=0.01,
    fill_mode='constant',
    cval = 0)
datagen2 = ImageDataGenerator(
    rotation_range=0,
    width_shift_range=0.18,
    height_shift_range=0,
    zoom_range=0.01,
    fill_mode='constant',
    cval = 0)
# Fit the ImageDataGenerator on the training data
datagen1.fit(x_train)
datagen2.fit(x_train)
# Generate augmented data
times = 8
x_augmented_post = np.concatenate((x_train, np.zeros((60000*times,28,28)).reshape(60000*times, 28, 28, 1)))
y_augmented_post = np.concatenate((y_train,y_train,y_train,y_train,y_train,y_train,y_train,y_train,y_train))

for j in range(times):
    for i in range(x_train.shape[0]):
        if(y_train['label'][i]!=0 and y_train['label'][i]!=1 and y_train['label'][i]!=2 and y_train['label'][i]!=3 and y_train['label'][i]!=4 and y_train['label'][i]!=6 and y_train['label'][i]!=8 ):
            x_augmented_post[i+j*60000] = datagen2.random_transform(x_train[i])
        else:
            x_augmented_post[i+j*60000] = datagen1.random_transform(x_train[i])
permutation = np.random.permutation(x_augmented_post.shape[0])
x_augmented_post = x_augmented_post[permutation]
y_augmented_post = y_augmented_post[permutation] 
cant_img = x_augmented_post.size/x_augmented_post[0].size
```

```python
c=[]
for k in range(x_augmented_post.shape[0]):
    if x_augmented_post[k].mean() < 1e-6:
        c.append(k)

k = np.delete(x_augmented_post,c, axis=0)
y = np.delete(y_augmented_post,c,axis = 0)
print(-k.shape[0]+x_augmented_post.shape[0])
```

```python
plt.figure(figsize=(12, 8))
#Mean and Variance of all clases.
for i in range(1,11):
    plt.subplot(5,4,2*i - 1)
    a = plt.imshow(k[np.random.randint(100000)], cmap='gray')
    #plt.colorbar(a)
    plt.subplot(5,4,2*i)
    b = plt.imshow(k[np.random.randint(100000)], cmap='gray')
    #plt.colorbar(b)
```

```python
X_train_par, X_val_par, y_train_par, y_val_par = train_test_split(k, y, test_size=0.4, random_state=42,shuffle = True)
```

```python
y_sparse_train = np.zeros([y_train_par.size,10])
y_sparse_val = np.zeros([y_val_par.size,10])
for k in range(y_train_par.size):
    y_sparse_train[k,y_train_par[k][0]] = 1
for k in range(y_val_par.size):
    #y_val['label'].values[k]
    y_sparse_val[k,y_val_par[k][0]] = 1
```

```python
from keras.layers import Dense, Flatten,Conv2D, MaxPooling2D,BatchNormalization,Dropout
from keras.optimizers import SGD, Adam, Nadam
from keras.models import Sequential
from keras.constraints import max_norm
from keras.callbacks import EarlyStopping, LearningRateScheduler,ReduceLROnPlateau
from keras.regularizers import l2
import visualkeras
```

```python
model = Sequential()
```

```python
#Competition

model.add(Conv2D(128, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(512, activation='relu',kernel_initializer='glorot_uniform'))
model.add(Dropout(0.2))
model.add(Dense(1024, activation='relu',kernel_initializer='glorot_uniform'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax',kernel_initializer='glorot_uniform'))
```

```python
#Tp
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten(input_shape=(28, 28, 1)))
model.add(Dense(256, activation='relu',kernel_initializer='glorot_normal'))
model.add(Dropout(0.4))
model.add(Dense(1024, activation='relu',kernel_initializer='glorot_normal'))
model.add(Dropout(0.3))
model.add(Dense(10, activation='softmax',kernel_initializer='glorot_normal'))
```

```python
model.summary()
```

```python
visualkeras.layered_view(model)
```

```python
 
model.compile(loss = "categorical_crossentropy",optimizer=Nadam(learning_rate=0.001,beta_1=0.999,decay=0,epsilon=1e-8), metrics = ["accuracy"])
#model.compile(loss = "categorical_crossentropy", optimizer = SGD(learning_rate = 0.001),metrics = ["accuracy"])
```

```python
reduceLR=ReduceLROnPlateau(monitor='val_loss', 
                           factor=0.2, 
                           patience=5, 
                           verbose=1, 
                           min_delta=0, 
                           cooldown=0, min_lr=0)
```

```python

callback = EarlyStopping(monitor='val_loss',min_delta=0, patience=30,restore_best_weights=True)
history = model.fit(X_train_par, y_sparse_train,
                  batch_size=256,
                  epochs=200,
                  verbose=1,
                  validation_data=(X_val_par, y_sparse_val),
                  callbacks=[callback,reduceLR])
```

```python
import seaborn as sns
epoch_count = range(1,len(history.history['accuracy']) + 1)
sns.lineplot(x=epoch_count, y = history.history['val_loss'],label = 'validation loss')
sns.lineplot(x=epoch_count, y = history.history['val_accuracy'],label = 'validation acc')
sns.lineplot(x=epoch_count, y = history.history['loss'],label = 'train loss')
sns.lineplot(x=epoch_count, y = history.history['accuracy'],label = 'train acc')
sns.lineplot(x=epoch_count, y = history.history['lr'],label = 'train acc')
plt.grid()
plt.show()
```
