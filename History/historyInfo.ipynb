```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
pd1 = pd.DataFrame(pd.read_csv("RS_L_0.csv"))
pd2 = pd.DataFrame(pd.read_csv("RS_L_1.csv"))
pd3 = pd.DataFrame(pd.read_csv("RS_L_2.csv"))
pd4 = pd.DataFrame(pd.read_csv("RS_L_3.csv"))
pd5 = pd.DataFrame(pd.read_csv("RS_L_4.csv"))
```

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
epoch_count = range(1,len(pd1['accuracy']) + 1)
sns.lineplot(x=epoch_count, y = pd1['val_accuracy'].values,label = 'Cin = 128,  F1-score = %0.4f Recall = %0.4f Precision = %0.4f AUC = %0.4f' %(pd1['f1'].values[0] ,pd1['recall'].values[0],pd1['precision'].values[0],pd1['auc'].values[0]))
sns.lineplot(x=epoch_count, y = pd2['val_accuracy'].values,label = 'Cin = 256, F1-score = %0.4f Recall = %0.4f Precision = %0.4f AUC = %0.4f' %(pd2['f1'].values[0] ,pd2['recall'].values[0],pd2['precision'].values[0],pd2['auc'].values[0]))
sns.lineplot(x=epoch_count, y = pd3['val_accuracy'].values,label = 'Cin = 512,  F1-score = %0.4f Recall = %0.4f Precision = %0.4f AUC = %0.4f' %(pd3['f1'].values[0] ,pd3['recall'].values[0],pd3['precision'].values[0],pd3['auc'].values[0]))
sns.lineplot(x=epoch_count, y = pd4['val_accuracy'].values,label = 'Cin = 1024, F1-score = %0.4f Recall = %0.4f Precision = %0.4f AUC = %0.4f' %(pd4['f1'].values[0] ,pd4['recall'].values[0],pd4['precision'].values[0],pd4['auc'].values[0]))
sns.lineplot(x=epoch_count, y = pd5['val_accuracy'].values,label = 'Cin = 2048,  F1-score = %0.4f Recall = %0.4f Precision = %0.4f AUC = %0.4f' %(pd5['f1'].values[0] ,pd5['recall'].values[0],pd5['precision'].values[0],pd5['auc'].values[0]))
plt.title("Regresion Softmax")
plt.xlabel("Epoch")
plt.ylabel("validation accuracy")
plt.grid()
plt.show()
```
